````python
from traitlets.config import Config
from nbconvert import MarkdownExporter, NbConvertBase
from datetime import datetime
import re

def make_codeblocks_collapsible(body):
    """
    Convert codeblocks to collapsible codeblocks. Combine adjacent codeblocks into one.
    """
    pattern = re.compile(r'```python\n(.*?)```\n', re.DOTALL)
    result = re.sub(
        pattern, r'<details>\n<summary>Code</summary>\n\n```Python\n\1```\n</details>', body)
    result = re.sub(r'</details>\s*<details>\n<summary>Code</summary>', '', result)
    return result

def export_notebook_to_md(filename, output_filename=None, append_version=True):
    c = Config()
    c.ExtractOutputPreprocessor.enabled = True
    c.ExtractOutputPreprocessor.extract_output_types = ['image/png', 'image/jpeg']

    exporter = MarkdownExporter(config=c)
    nb_converter = NbConvertBase(config=c)
    notebook, resources = nb_converter.from_filename(filename)
    (body, _) = exporter.from_notebook_node(notebook, resources=resources)

    if append_version:
        output_filename = output_filename.split('.')[0] + '_' + datetime.now().strftime("%Y-%m-%d_%H%M") + '.md'

    body = make_codeblocks_collapsible(body)
    result = {
        'body': body,
        'metadata': resources['metadata']
    }

    if output_filename:
        with open(output_filename, 'w', encoding='utf-8') as f:
            f.write(body)
        print(f'File saved to {output_filename}')

        # Save extracted figures
        for output in resources['outputs']:
            if output.output_type.startswith('image/') and 'data' in output:
                image_data = output.data
                output_filename = output.metadata.get('filename')
                if output_filename:
                    with open(output_filename, 'wb') as f:
                        f.write(image_data)

    return result

file = '2023-07-04 portfolio notebook.ipynb'
response = export_notebook_to_md(file, 'README.md')
````

## 6.0 create_readme

````python
from traitlets.config import Config
from nbconvert import MarkdownExporter
from nbconvert.preprocessors import ExtractOutputPreprocessor
from datetime import datetime
import re

def make_codeblocks_collapsible(body):
    """
    Convert codeblocks to collapsible codeblocks. Combine adjacent codeblocks into one.
    """
    pattern = re.compile(r'```python\n(.*?)```\n', re.DOTALL)
    result =  re.sub(
        pattern, r'<details>\n<summary>Code</summary>\n\n```Python\n\1```\n</details>', body)
    result = re.sub(r'</details>\s*<details>\n<summary>Code</summary>', '', result)
    return result

def export_notebook_to_md(filename, output_filename=None, figure_path='../outputs/figures/', append_version=True):
    c = Config()
    c.ExtractOutputPreprocessor.enabled = True
    
    # Create the ExtractOutputPreprocessor with the given configuration
    preprocessor = ExtractOutputPreprocessor(config=c)
    
    exporter = MarkdownExporter(config=c)
    (body, resources) = exporter.from_filename(filename, resources={"metadata": {"path": ""}}, preprocessors=[preprocessor])

    if append_version:
        output_filename = output_filename.split('.')[0] + '_' + datetime.now().strftime("%Y-%m-%d_%H%M") + '.md'

    body = make_codeblocks_collapsible(body)
    figure_path_from_root = figure_path.split('../')[-1]
    body = re.sub(r'!\[(.*?)\]\((.*?)\)', r'![\1]('+figure_path_from_root+r'\2)', body)
    result = {
        'body': body,
        'response': resources
    }
    if output_filename:
        with open(output_filename, 'w', encoding='utf-8') as f:
            f.write(body)
        print(f'Markdown file saved to {output_filename}')

        # Save extracted figures
        for output_entry in resources['outputs']: 
            filename = figure_path+output_entry    
            with open(filename , 'wb') as f:
                f.write(resources['outputs'][output_entry])
            print(f'Saving {filename}')

    return result

def create_readme(filename, output_filename='README.md', append_version=False):
    output_filename=f'../{output_filename}'

    result = export_notebook_to_md(
        filename, output_filename, append_version=append_version)
    return result

file = '2023-07-04 portfolio notebook.ipynb'
# response = export_notebook_to_md(file, 'README.md')

result = create_readme(file)
````
